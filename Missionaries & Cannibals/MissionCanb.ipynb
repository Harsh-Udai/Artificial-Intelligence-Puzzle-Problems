```python
# Contributor: Harsh Udai :)

# Missionaries and Cannibals Puzzle Problem
# Approch-------
# 1. General state creation with the contraints.
# 2. Graph Creation from states.
# 3. States of both left and right side are cosidered.
# 3. Depth First Search used to find the path.
```

```python
class MissionCaneb:
    def __init__(self,a,b):
        self.init_X=a
        self.init_Y=b
        self.fin_X=0
        self.fin_Y=0
        self.graph={}
        self.final=[]
        
    def state_S(self,a,b,c,d):
        l=[]
        if(self.error_RC(a-1,b-1,c+1,d+1)):
#             print(a-1,b-1," -> ",c+1,d+1)
            l.append([a-1,b-1,c+1,d+1])
            
        if(self.error_RC(a-2,b,c+2,d)):
#             print(a-2,b," -> ",c+2,d)
            l.append([a-2,b,c+2,d])
            
        if(self.error_RC(a,b-2,c,d+2)):
#             print(a,b-2," -> ",c,d+2)
            l.append([a,b-2,c,d+2])
            
        return l
            
    def error_RC(self,a,b,c,d):
        if((b>a and a!=0) or (d>c and c!=0) or (a<0 or b<0 or c<0 or d<0)):
            return 0
        else:
            return 1
        
    def return_State(self,a,b,c,d):
        l=[]
        
        if(self.error_RC(a+1,b+1,c-1,d-1)):
#             print(a+1,b+1," -> ",c-1,d-1)
            l.append([a+1,b+1,c-1,d-1])
            
        if(self.error_RC(a+1,b,c-1,d)):
#             print(a+1,b," -> ",c-1,d)
            l.append([a+1,b,c-1,d])
        
        if(self.error_RC(a,b+1,c,d-1)):
#             print(a,b+1," -> ",c,d-1)
            l.append([a,b+1,c,d-1])
        
        if(self.error_RC(a+2,b,c-2,d)):
#             print(a+2,b," -> ",c-2,d)
            l.append([a+2,b,c-2,d])
            
        if(self.error_RC(a,b+2,c,d-2)):
#             print(a,b+2," -> ",c,d-2)
            l.append([a,b+2,c,d-2])
            
        return l
        
    def addEdge(self,a,b,c,d,p,a1,b1,c1,d1,p1):
        
        if(self.graph.get((a,b,c,d,p))==None):
            self.graph[(a,b,c,d,p)]=[]
            self.graph[(a,b,c,d,p)].append((a1,b1,c1,d1,p1))
        else:
            self.graph[(a,b,c,d,p)].append((a1,b1,c1,d1,p1))
            
    def showGraph(self):
        print(self.graph)
        
    # DFS Technique
    def DFSUtil(self, u,v,u1,v1,p, visited,l):

        visited.add((u,v,u1,v1,p))
        l.append((u,v,u1,v1,p))
        print(u,v,u1,v1,p, end='  ')
      
        if((u,v,u1,v1)==(self.fin_X,self.fin_Y,self.init_X,self.init_Y)):
            # print("mil gaya")
            self.final = [i for i in l]
        
        if(self.graph.get((u,v,u1,v1,p))!=None):
            for neighbour in self.graph[(u,v,u1,v1,p)]:
                print("----------NBH",neighbour)
                if neighbour not in visited:
                    self.DFSUtil(neighbour[0],neighbour[1],neighbour[2],neighbour[3],neighbour[4], visited,l)
                    
        else:
            l.pop()
            
    def DFS(self):
        u,v,u1,v1,p=self.init_X,self.init_Y,self.fin_X,self.fin_Y,'l'
        visited = set()
        l=[]
        self.DFSUtil(u,v,u1,v1,p, visited,l)
            
    def path(self):
        return self.final
        
    def stateFinder(self):
        
        duplicate_check=[]
        state_change=[]
        state_transfer=[]
        
        left_sider=[]
        right_sider=[]
        
        a,b,c,d=self.init_X,self.init_Y,self.fin_X,self.fin_Y
        state_change.append((a,b,c,d))
        duplicate_check.append((a,b,c,d))
        
        while(len(state_change)!=0):
            
            print(a,b,c,d)
            ans = self.state_S(a,b,c,d)
            print("Possible Nodes",ans)

            if(len(ans)==0 and len(state_change)==0):
                break
                
            elif(len(state_change)!=0 and len(ans)==0):
                state_change.pop(0)
                while(True and len(state_change)!=0):
                    a,b,c,d=state_change[0]
                    ans = self.state_S(a,b,c,d)
                    if(len(ans)!=0 or len(state_change)==0):
                        break
                    else:
                        state_change.pop(0)
                    print("Possible Nodes",ans)
            
            if(len(state_change)==0):
                break
                
            state_transfer.append(state_change[0])
            state_change.pop(0)
            left_sider.append((a,b,c,d))
            
            for i in ans:
                
                self.addEdge(a,b,c,d,'l',i[0],i[1],i[2],i[3],'r')
                
                if ((i[0],i[1],i[2],i[3]) not in duplicate_check):
                    duplicate_check.append((i[0],i[1],i[2],i[3]))
                    right_sider.append((i[0],i[1],i[2],i[3]))
                    
                print("Possible return edge",self.return_State(i[0],i[1],i[2],i[3]))
                
                for j in self.return_State(i[0],i[1],i[2],i[3]):
                    if((j[0],j[1],j[2],j[3])!=(a,b,c,d) and (j[0],j[1],j[2],j[3]) not in state_transfer and 
                      (j[0],j[1],j[2],j[3]) not in state_change ):
                        state_change.append((j[0],j[1],j[2],j[3]))
                        self.addEdge(i[0],i[1],i[2],i[3],'r',j[0],j[1],j[2],j[3],'l')
                        
                        
            
            print("state Change",state_change)
            # print("State Transfer",state_transfer)
            # print("Left sider",left_sider)
            # print("Right_sider",right_sider)
            # print("Graph",self.graph)

            if(len(state_change)==0):
                break

            print("----------------------------------","done") 
            a,b,c,d = state_change[0]
```
